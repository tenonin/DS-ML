```python
df_test.drop(['UniqueCarrier','DepHour'],axis=1,inplace=True)
df_train.drop(['UniqueCarrier','DepHour'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_train = df_train.drop('dep_delayed_15min',axis=1)
y_train = df_train.loc[:,'dep_delayed_15min']
X_test = df_test
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('scaler', StandardScaler()), ('GBTC', GradientBoostingClassifier())])
pipeline
```

```python
from sklearn.model_selection import GridSearchCV

parameteres = {'GBTC__criterion':['friedman_mse'],
               'GBTC__n_estimators':range(1000,1600,200),
               'GBTC__min_samples_split':range(500,1600,500),
               'GBTC__min_samples_leaf':[30,60,90],
               'GBTC__warm_start':[True],
               'GBTC__n_iter_no_change':[15],
               'GBTC__max_depth':[6,8,10]}

grid = GridSearchCV(pipeline, param_grid=parameteres,cv=5)
```

```python
%%time
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict_proba(X_test)
y_pred
```
