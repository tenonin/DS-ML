```python
import pandas as pd
```

```python
df = pd.read_csv('bank.csv')
df
```

```python
df.info()
```

```python
def convert_dict(column):
    dic = {}
    lista = df[column].value_counts().index.tolist()
    
    for i in range(len(lista)):
        dic[lista[i]] = i
        
    return dic
```

```python
job = convert_dict('job')
marital = convert_dict('marital')
education = convert_dict('education')
housing = convert_dict('housing')
default = convert_dict('default')
loan = convert_dict('loan')
contact = convert_dict('contact')
month = convert_dict('month')
poutcome = convert_dict('poutcome')
deposit = convert_dict('deposit')

df['marital'] = df['marital'].map(marital)
df['job'] = df['job'].map(job)
df['education'] = df['education'].map(education)
df['housing'] = df['housing'].map(housing)
df['default'] = df['default'].map(default)
df['loan'] = df['loan'].map(loan)
df['contact'] = df['contact'].map(contact)
df['month'] = df['month'].map(month)
df['poutcome'] = df['poutcome'].map(poutcome)
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df['deposit'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('deposit',axis=1)
y = df.loc[:,'deposit']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('scaler', StandardScaler()), ('MLP', MLPClassifier())])
pipeline
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'MLP__hidden_layer_sizes':[50,100,200],
              'MLP__activation':['relu','identity'],
              'MLP__learning_rate_init':[0.001],
              'MLP__warm_start':[True],
              'MLP__max_iter':[1000]}

grid = GridSearchCV(pipeline, param_grid=parameters)
```

```python
%%time
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(matrix,annot=True,fmt="d",cmap='Blues')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
