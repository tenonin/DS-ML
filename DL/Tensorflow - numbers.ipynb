```python
import tensorflow as tf
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(2, 2), dpi=200)

for i in range(4):
    print('Value of {} is {}'.format(i,y_train[i]))
    fig.add_subplot(2,2,i+1)
    plt.imshow(X_train[i])
plt.show()
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.output_shape
```

```python
history = model.fit(X_train, y_train,epochs=50, verbose=1)
```

```python
history.history.keys()
```

```python
import numpy as np

plt.plot(np.arange(50), history.history['loss'], '-')
plt.title('Loss Function')
```

```python
plt.plot(np.arange(50), history.history['accuracy'], '-')
plt.title('Accuracy')
```

```python
evaluation = model.evaluate(X_test, y_test, verbose = 1)
```

```python
print('Accuracy of model: {}'.format(evaluation[1]))
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(2, 2), dpi=200)

print("NOTE:if value predicted is nearly 1.0, it's the true value\n\n\n")
for i in range(16):
    print('Value of {} is {}. Predticted {}'.format(i,y_test[i],y_pred[i][y_test[i]]))
    fig.add_subplot(4,4,i+1)
    plt.imshow(X_test[i])
plt.show()
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.output_shape
```

```python
history = model.fit(X_train, y_train,epochs=50, verbose=1, batch_size=150)
```

```python
evaluation = model.evaluate(X_test, y_test, verbose = 1)
```

```python
import numpy as np

plt.plot(np.arange(50), history.history['loss'], '-')
plt.title('Loss Function')
```

```python
plt.plot(np.arange(50), history.history['accuracy'], '-')
plt.title('Accuracy')
```

```python
print('Accuracy of model: {}'.format(evaluation[1]))
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(2, 2), dpi=200)

print("NOTE:if value predicted is nearly 1.0, it's the true value\n\n\n")
for i in range(16):
    print('Value of {} is {}. Predticted {}'.format(i,y_test[i],y_pred[i][y_test[i]]))
    fig.add_subplot(4,4,i+1)
    plt.imshow(X_test[i])
plt.show()
```
